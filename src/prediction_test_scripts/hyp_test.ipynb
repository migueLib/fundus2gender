```python
import cv2
import os
from random import sample
import pandas as pd
```

```python
input_dir = './highscore_pictures/'

female = os.listdir(input_dir + 'female/')
male = os.listdir(input_dir + 'male/')

female = list(((i,0) for i in female))
male = list(((i,1) for i in male))

combined = female + male

dfc = pd.DataFrame(combined, columns=["picture", "True Label"])

len(female), len(male)
```

```python
dfc = dfc.replace({0:'female',1:'male'})

dfc.head()
```

```python
combined_test = sample(combined, 50)
mode="Train" #"Test"
labeler="bug_test2"
```

```python
correct = 0
total = 0
current = 0
log = []
quit = False
folder = ""
labeled = False

cv2.namedWindow("Manual Classification")

for i, l in combined_test:
    
    if(l == 0):
        folder = 'female/'
    else:
        folder = 'male/'
    
    image = cv2.imread(input_dir + folder + i)
    
    h,w,_ = image.shape
    labeled = False
    
    while True:
        cv2.imshow("Manual Classification", image)
        key = cv2.waitKey(1) & 0xFF
        
        if key == ord("f") and not labeled:
            current = 0
            total += 1
            log.append((i,'female'))
            if(current == l):
                correct += 1
            if(mode=="Test"):
                break;
            else:
                labeled = True
                if(current == l):
                    image = cv2.circle(image, ((int) (w/2), (int) (h/2)), ((int) (h/6)), (0,255,0), ((int) (h/24)))
                else:
                    image = cv2.drawMarker(image, ((int) (w/2), (int) (h/2)), (0,0,255), cv2.MARKER_TILTED_CROSS, thickness=32, markerSize=((int) (h/6)))
            
        if key == ord("m") and not labeled:
            current = 1
            total += 1
            log.append((i,'male'))
            if(current == l):
                correct += 1
            if(mode=="Test"):
                break;
            else:
                labeled = True
                if(current == l):
                    image = cv2.circle(image, ((int) (w/2), (int) (h/2)), ((int) (h/6)), (0,255,0), ((int) (h/24)))
                else:
                    image = cv2.drawMarker(image, ((int) (w/2), (int) (h/2)), (0,0,255), cv2.MARKER_TILTED_CROSS, thickness=32, markerSize=((int) (h/6)))
                    
        if key == 32:
            break;
            
        if key == ord("q"):
            quit = True
            break;
            
    if(quit):
        break;
        
cv2.destroyAllWindows()

if(mode == "Test"):
    dfa = pd.DataFrame(log, columns=["picture",labeler])
    dfc[labeler] = dfc['picture'].map(dfa.set_index('picture')[labeler])
        
print("Classified {} out of {} correctly. Accuracy of {} reached.".format(correct, total, correct/total))
```
