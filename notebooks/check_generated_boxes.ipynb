```python
def unshrink(ls):
    nls = []
    for i, v in enumerate(ls):
        v = v * h/299
        if i % 2 == 0:
            v = v + ((w-h)/2)
        nls.append(v)
    return nls
```

```python
df['true'] = df['true'].apply(unshrink)
df['predicted'] = df['predicted'].apply(unshrink)
```

```python
true_coords = df['true'].apply(pd.Series)
predicted_coords = df['predicted'].apply(pd.Series)

true_coords = true_coords.rename(columns={0: "t_OD_UL_x", 1:"t_OD_UL_y", 2: "t_OD_LR_x", 3:"t_OD_LR_y", 4: "t_FV_UL_x", 5:"t_FV_UL_y", 6: "t_FV_LR_x", 7:"t_FV_LR_y"})
predicted_coords = predicted_coords.rename(columns={0: "p_OD_UL_x", 1:"p_OD_UL_y", 2: "p_OD_LR_x", 3:"p_OD_LR_y", 4: "p_FV_UL_x", 5:"p_FV_UL_y", 6: "p_FV_LR_x", 7:"p_FV_LR_y"})
true_coords = true_coords.round(0).astype(int)
predicted_coords = predicted_coords.round(0).astype(int)

dfs = true_coords.join(predicted_coords)

dfs
```

```python
dfs["t_OD_area"] = (dfs["t_OD_LR_x"] - dfs["t_OD_UL_x"]) * (dfs["t_OD_LR_y"] - dfs["t_OD_UL_y"])
dfs["t_FV_area"] = (dfs["t_FV_LR_x"] - dfs["t_FV_UL_x"]) * (dfs["t_FV_LR_y"] - dfs["t_FV_UL_y"])
dfs["p_OD_area"] = (dfs["p_OD_LR_x"] - dfs["p_OD_UL_x"]) * (dfs["p_OD_LR_y"] - dfs["p_OD_UL_y"])
dfs["p_FV_area"] = (dfs["p_FV_LR_x"] - dfs["p_FV_UL_x"]) * (dfs["p_FV_LR_y"] - dfs["p_FV_UL_y"])

dfs["t_p_OD_intersect"] = ( (dfs[["t_OD_LR_x","p_OD_LR_x"]].min(axis = 1) - dfs[["t_OD_UL_x","p_OD_UL_x"]].max(axis = 1)).clip(lower=0) 
                            * (dfs[["t_OD_LR_y","p_OD_LR_y"]].min(axis = 1) - dfs[["t_OD_UL_y","p_OD_UL_y"]].max(axis = 1)).clip(lower=0) )
dfs["t_p_FV_intersect"] = ( (dfs[["t_FV_LR_x","p_FV_LR_x"]].min(axis = 1) - dfs[["t_FV_UL_x","p_FV_UL_x"]].max(axis = 1)).clip(lower=0) 
                            * (dfs[["t_FV_LR_y","p_FV_LR_y"]].min(axis = 1) - dfs[["t_FV_UL_y","p_FV_UL_y"]].max(axis = 1)).clip(lower=0) )

dfs["t_p_OD_union"] = dfs["t_OD_area"] + dfs["p_OD_area"] - dfs["t_p_OD_intersect"]
dfs["t_p_FV_union"] = dfs["t_FV_area"] + dfs["p_FV_area"] - dfs["t_p_FV_intersect"]

dfs["Jaccard_Index_OD"] = dfs["t_p_OD_intersect"] / dfs["t_p_OD_union"]
dfs["Jaccard_Index_FV"] = dfs["t_p_FV_intersect"] / dfs["t_p_FV_union"]
```
